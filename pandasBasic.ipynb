```python
import pandas as pd
import seaborn as sns
```

```python
#Load the tips dataset
df = sns.load_dataset('tips')

# Quick preview
df.head()
```

```python
# show structure of the dataframe

df.info()
```

```python
# get summary statistics

df.describe()
```

```python
df.isnull().sum()
```

```python
#Value count for a categorical column

df['day'].value_counts()
```

```python
df['time'].value_counts()
```

```python
# Average tip amoung

df['tip'].mean()
```

```python
# Correlation between tip and total_bill

df[['tip', 'total_bill']].corr()
```

```python
# filter tips for more than 5

df[df['tip'] > 5]
```

```python
# Create a new column tip percentage

df['tip_pct'] = df['tip'] / df['total_bill']
df.head()
```

```python
#Group by day, get average tip

df.groupby('day')['tip'].mean()
```

```python
#Sort by biggest tip 

df.sort_values('tip', ascending=False).head(10)
```

```python
import matplotlib.pyplot as plt

sns.scatterplot(data=df, x = 'total_bill', y = 'tip', hue = 'time')
plt.title('Total Bill vs Tip by Time')
plt.show()
```
